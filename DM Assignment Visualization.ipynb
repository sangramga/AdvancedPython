```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
# Read the whole text.
text = open("/home/sangram/AdvancedPython/cluster.txt").read()
# Generate a word cloud image
wordcloud = WordCloud(max_font_size=40,width=600,height=400,relative_scaling=0.5).generate(text)
# Display the generated image:
# the matplotlib way:


plt.figure(figsize=(25, 10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
%matplotlib inline
import nltk
from nltk.corpus import stopwords


stopWords = set(stopwords.words('english'))

f = open("/home/sangram/AdvancedPython/cluster.txt", "r")
inputfile = f.read()
tokens = nltk.tokenize.word_tokenize(inputfile)
tokens=[token.lower() for token in tokens if token.isalpha()] 
tokens = [token for token in tokens if token not in stopWords]
fd = nltk.FreqDist(tokens)
print(type(fd.values()))
plt.figure(figsize=(25, 10))
fd.plot(50,cumulative=False)

```

```python
from nltk.util import ngrams
from collections import Counter
import pandas as pd
bigrams = ngrams(tokens,3)
fd = nltk.FreqDist(bigrams)
plt.figure(figsize=(25, 10))
plt.xlabel('samples', fontsize=18)
plt.ylabel('counts', fontsize=16)
fd.plot(70,cumulative=False)
# frequencies = Counter(bigrams)
# print(frequencies[('fidel','castro')])
# bigram_words = frequencies.keys()
# counts = frequencies.values()
# print(bigram_words)
```
