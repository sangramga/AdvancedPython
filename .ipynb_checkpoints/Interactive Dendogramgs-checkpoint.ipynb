# Non Interactive plot

```python
"""
=========================================
Plot Hierarachical Clustering Dendrogram 
=========================================
This example plots the corresponding dendrogram of a hierarchical clustering
using AgglomerativeClustering and the dendrogram method available in scipy.
"""

import numpy as np

from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering

def plot_dendrogram(model, **kwargs):

    # Children of hierarchical clustering
    children = model.children_

    # Distances between each pair of children
    # Since we don't have this information, we can use a uniform one for plotting
    distance = np.arange(children.shape[0])

    # The number of observations contained in each cluster level
    no_of_observations = np.arange(2, children.shape[0]+2)

    # Create linkage matrix and then plot the dendrogram
    linkage_matrix = np.column_stack([children, distance, no_of_observations]).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)


iris = load_iris()
x = iris.data[:20]
model = AgglomerativeClustering(n_clusters=3)

model = model.fit(x)
plt.rcParams["figure.figsize"] = (20,6)
plt.title('Hierarchical Clustering Dendrogram')
plot_dendrogram(model, labels=model.labels_)
plt.show()
```

# Interactive Dendograms plotting

```python
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
from IPython.display import display
%matplotlib inline
```

```python


def cluster_and_plot(n_clusters):
    iris = load_iris()
    x = iris.data[:30]
    model = AgglomerativeClustering(n_clusters=n_clusters)

    model = model.fit(x)
    plt.rcParams["figure.figsize"] = (20,6)
    plt.title('Hierarchical Clustering Dendrogram')
    plot_dendrogram(model, labels=model.labels_)
    plt.show()

interactive_plot = interactive(cluster_and_plot, n_clusters=(2,5,1))
# output = interactive_plot.children[-1]
# output.layout.height = '350px'
interactive_plot
```

# DM Assignment Dendogram

```python
data = open("/home/sangram/AdvancedPython/cluster.txt").read().split('\n')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer(max_df=0.9,min_df=0.05,analyzer='word',stop_words='english')
count_matrix = count_vect.fit_transform(data)
```

```python
count_matrix = count_matrix.toarray()
def cluster_and_plot1(n_clusters):
    iris = load_iris()
    small_data = count_matrix[:500,:]
    model = AgglomerativeClustering(n_clusters=n_clusters,affinity='cosine',linkage='complete')

    model = model.fit(small_data)
    plt.rcParams["figure.figsize"] = (20,10)
    plt.title('Hierarchical Clustering Dendrogram')
    plot_dendrogram(model, labels=model.labels_)
    plt.show()

interactive_plot = interactive(cluster_and_plot1, n_clusters=(20,25,1))
# output = interactive_plot.children[-1]
# output.layout.height = '350px'
interactive_plot
```

```python
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
numpy_matrix = count_matrix.toarray()
# print(count_matrix.dtype)
# print(count_matrix.shape)
Z = linkage(numpy_matrix,'single')
```

```python
# calculate full dendrogram
plt.figure(figsize=(25, 10))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('sample index')
plt.ylabel('distance')
dendrogram(
    Z,
    leaf_rotation=90.,  # rotates the x axis labels
    leaf_font_size=8.,  # font size for the x axis labels
)
plt.show()
```
